# 2. Array manipulation functions

```python
# reshape an array
a1 = np.random.randint(0,500,(3,4,5))
print('Original array: ',a1)
print('Reshaped array: ',a1.reshape((6,2,5)))
```

```python
#flatten an array Flattened to 1d array
a2 = np.random.randint(0,100,(5,7))
print('Original array: ',a2)
print('flattened array: ',np.ravel(a2))
```

```python
#transpose of an array
a3 = np.random.randint(0,100,(2,6))
print('Original array: ',a3)
print('Transposed array: ',np.transpose(a3))
```

```python
#stack an array vertically
a4 = np.random.randint(0,600,(2,6))
a5 = np.random.randint(0,400,(2,6))
a6 = np.vstack([a4,a5])
print('stacked array: ',a6)
```

```python
#stack an array horizontally
a4 = np.random.randint(0,600,(2,6))
a5 = np.random.randint(0,400,(2,6))
a6 = np.hstack([a4,a5])
print('stacked array: ',a6)
```

# 3. Mathematical functions

```python
# add 2 to all elements
am1 = np.arange(5)
print('new array: ',am1+2) # np.add(am1,2)
```

```python
# Square each element
am2 = np.arange(0,11)
print("Squared g:", am2**2)
```

```python
# square root of n array
am3 = np.arange(0,11)
print('new array: ',np.sqrt(am3)) 
```

```python
# dot product of an array
am4 = np.arange(0,11)
am5 = np.arange(33,44)
print('new array: ',np.dot(am4,am5)) 
```

```python
# Calculate the dot product  
dot_product = np.dot(am4, am5)  
print('Dot Product:', dot_product)  

# Calculate the magnitudes (norms) of the vectors  
magnitude_am4 = np.linalg.norm(am4)  
magnitude_am5 = np.linalg.norm(am5)  

# Calculate the cosine of the angle  
cos_theta = dot_product / (magnitude_am4 * magnitude_am5)  

# Calculate the angle in radians  
angle_radians = np.arccos(cos_theta)  

# Convert the angle to degrees  
angle_degrees = np.degrees(angle_radians)  

print('Angle (in radians):', angle_radians)  
print('Angle (in degrees):', angle_degrees)
```

# 4. Statistical Functions

```python
as1 = np.arange(20)
print('Array is:', as1)
print('Array mean is:', np.mean(as1))
print('Array median is:', np.median(as1))
```

```python
as2 = np.random.randint(1,7,20000)  # rolling a die 20000 times and check the result and have an avg outcome along with deviation
print('Array is:', as2)
print('Array sum is:', np.sum(as2))
print('Array mean is:', np.mean(as2))
print('Array median is:', np.median(as2))
print('Array standard deviation is:', np.std(as2))
print('Array variance is:', np.var(as2))
```

```python
# Central limit theorm
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns

sum=[]
for i in range(5000):
    as2 = np.random.randint(1,7,20000)  # rolling a die 20000 times and check the result and have an avg outcome along with deviation
    sum.append(np.sum(as2))
df = pd.DataFrame(sum)

# draw the histplot
sns.histplot(df, kde=True)
```
