# 1. Array creation functions

```python
import numpy as np
```

```python
# CReating an array from a list
a = [1,2,3]
arr = np.array(a)
print('Array is:', arr)
```

```python
# create an array with evenly spaced values
arr1 = np.arange(0,10,2) # This create an array from 0 to 10 with a step 2
print('Array is:', arr1)
```

```python
# create an array with linearly spaced values
arr2 = np.linspace(0,1,5) # 5 values evenly spaced b/w 0 and 1
print('Array is:', arr2)
```

```python
# create an 1D-array filled with zeros
arr3 = np.zeros(2)
print('1-D Array is:', arr3)
```

```python
# create a 2D-array filled with zeros
arr4 = np.zeros((2,3))
print('2-D Array is:', arr4)
```

```python
# create a 3D-array filled with zeros
arr5 = np.zeros((2,3,3))
print('3-D Array is:', arr5)
```

```python
# create a 3D-array filled with zeros with integers
arr6 = np.zeros((2,3,3),dtype = int)
print('3-D Array is:', arr6)
```

```python
type(arr5) == type(arr6)
```

```python
arr5 == arr6
```

```python
# create a 3D-array filled with ones
arr7 = np.ones((2,3,3))
print('3-D Array is:', arr7)
```

```python
# create a 3D-array filled with ones and integer
arr7 = np.ones((2,3,3),dtype = int)
print('3-D Array is:', arr7)
```

```python
# create an identity matrix
arr8 = np.identity(6)
print('identity matrix is :', arr8)
```

```python
# random values generation
# rand creates uniformly distributed values
arr9 = np.random.rand(2,7)
print('random vale array :', arr9)
```

```python
# to generate a uniformly distributed matrix with specified numbers and size
#np.random.uniform(low=0,high=1,size=(2,7))
arr10 = np.random.uniform(0,100,size=(2,7))
print('random vale array :', arr10)
```

```python
#randn creates standard normally distributed values
arr11 = np.random.randn(2,7)
print('random vale array :', arr11)
```

```python
# to generate a normal distributed matrix with specified numbers and size
# np.random.normal(loc=0,scale=1,size=(2,7)) loc = mean, scale = sd, size
arr12 = np.random.normal(loc=5,scale=1,size=(2,7))
print('random vale array :', arr12)
```

```python
#randint to generate random integers
#np.random.randint(low,high,size) 
arr13 = np.random.randint(0,100,(2,7))
print('random vale array :', arr13)
```

```python
#normallly distributed integer values
a= np.random.normal(loc=400,scale=400,size=(2,7))
# #uniformly distributed integer values
# a= np.random.uniform(low=1,high=4,size=(2,7))
print(a)
b = np.round(a).astype(int) #to makes the values integers
print(b)
c= np.clip(b,1,1000) # to clip the values b/w certain range
print(c)
```
