```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # we have to work with pyplot interface

%matplotlib inline
plt.rcParams['figure.figsize'] =6,4

import warnings
warnings.filterwarnings('ignore')
```

```python
%matplotlib inline
x1 = np.linspace(0,10,500)

#creata a plot figure
fig=plt.figure()

plt.plot(x1,np.sin(x1),'--')
plt.plot(x1,np.cos(x1),'-')
plt.show()
```

```python
plt.subplot(1,1,1)
plt.plot(x1,np.sin(x1),x1,np.cos(x1))
```

```python
plt.title('Sub plots 2*2')

plt.subplot(2,2,1)
plt.plot(x1,np.sin(x1),x1,np.cos(x1))
plt.axis([0,10,-1,1])
plt.xticks([2,4,6,8])
plt.yticks([-1, -0.5, 0, 0.5, 1])

plt.subplot(2,2,2)
plt.plot(x1,np.sin(x1),x1,np.cos(x1))
plt.axis([0,10,-1,1])
plt.xticks([2,4,6,8])
plt.yticks([-1, -0.5, 0, 0.5, 1])

plt.subplot(2, 2, 3)
plt.plot(x1,np.sin(x1),x1,np.cos(x1))
plt.axis([0,10,-1,1])
plt.xticks([2,4,6,8])
plt.yticks([-1, -0.5, 0, 0.5, 1])

plt.subplot(2,2,4)
plt.plot(x1,np.sin(x1),x1,np.cos(x1))
plt.axis([0,10,-1,1])
plt.xticks([2,4,6,8])
plt.yticks([-1, -0.5, 0, 0.5, 1])
```

```python
x= np.linspace(0,10,100)

plt.plot(x,x,label = 'Linear')
plt.plot(x,x**2,label = 'quadratic')
plt.plot(x,x**3,label = 'cubic')

# print title
plt.title('simple plots')

# print lables
plt.xlabel('X values')
plt.xlabel('y values')
#print x&y ticks
plt.xticks([i for i in range(11)])
plt.yticks([i*10 for i in range(11)])

#print axis values
plt.axis([0,10,0,100])

plt.legend()
```

```python
# formating style

x= np.linspace(0,10,100)

plt.plot(x,x,'go',label = 'Linear')
plt.plot(x,x**2,'h',label = 'quadratic')
plt.plot(x,x**3,'+',label = 'cubic')

# print title
plt.title('simple plots')

# print lables
plt.xlabel('X values')
plt.xlabel('y values')
#print x&y ticks
plt.xticks([i for i in range(11)])
plt.yticks([i*10 for i in range(11)])

#print axis values
plt.axis([0,10,0,100])

plt.legend()
```

```python
# ploting multiple graphs in single graph
t = np.arange(0,10,1)
plt.plot(t,t,'r+',t,t*2,'bo',t*1.5,t/3,'g^')
```

```python
# working with object oriented API
fig, ax = plt.subplots(2)

# calling plot() methods
ax[0].plot(t,t)
ax[0].set_title('Linear curve')
ax[0].set_xlabel('t')
ax[0].set_ylabel('t')
ax[1].plot(t,t**2)
ax[1].set_title('Parabolic curve')
ax[1].set_xlabel('t')
ax[1].set_ylabel('t^2')
```

```python
# objects and refrences
fig1 = plt.figure()
fig2 = plt.figure()

x1= np.linspace(0,100,10)
y1 = x1**2

axes1 = fig1.add_axes([0,0,0.5,0.5])
axes1.plot(x1, y1, 'r')
axes1.set_xlabel('x1')
axes1.set_ylabel('y1')
axes1.set_title('title 1');

axes2 = fig2.add_axes([0,0,0.5,0.5])
axes2.plot(x1*4, y1/5, 'y')
axes2.set_xlabel('x1*4')
axes2.set_ylabel('y1/5')
axes2.set_title('title 2')
```
